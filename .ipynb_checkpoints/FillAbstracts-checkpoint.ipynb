```python
# Load the database 
database_dict = pickle.load(open('data/updated_database_dict.p', 'rb'))
```

```python
for url_string in urls_list[-2:-1]: 
    abstract = database_dict[url_string]['abstract']
    if abstract is None: 
        print url_string
    Nkeywords = len(database_dict[url_string]['keyword_dict'])
    print database_dict[url_string]['keyword_dict']
    word_freq_list = [(entry['value'], Nkeywords - float(entry['rank'])) for entry in database_dict[url_string]['keyword_dict']]
        
    wordcloud = WordCloud().generate_from_frequencies(word_freq_list)
    plt.imshow(wordcloud)
```
